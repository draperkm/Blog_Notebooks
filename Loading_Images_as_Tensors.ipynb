# Loading images wit PyTorch: ImageFolder class

```python
%matplotlib inline
import matplotlib.pyplot as plt
import torch
from torchvision import datasets, transforms
```

When it comes to loading image data with PyTorch, the ImageFolder class works very nicely, and if you are planning on collecting the image data yourself, I would suggest organizing the data so it can be easily accessed using the ImageFolder class.

The easiest way to load image data is by using datasets.ImageFolder from torchvision. 

Here is how the ImageFolder class has to be organized to be able to upload the images:

```python
# dset.ImageFolder: it creates a list of directory names: A generic data loader where the images are arranged 
# in this way by default:

# root/dog/xxx.png
# root/dog/xxy.png
# root/dog/[...]/xxz.png
# root/cat/123.png
# root/cat/nsdf3.png
# root/cat/[...]/asd932_.png
```

We now define how we want to transoform the images uploades, by:

- Resizing them 
- Cropping them to the desidered final size (128 x 128 in this case)
- Converting them into tensors

```python
transform = transforms.Compose([transforms.Resize(150),
                                
                                transforms.CenterCrop(128), 
                                # size – Desired crop size. size is a sequence like (h, w), 
                                # where h and w are the height and width of the cropped image. 
                                # If size is anint, the cropped image will be a square image.
                                
                                transforms.ToTensor()])
```

```python
Let us now create the lis
```

```python
dataset = datasets.ImageFolder(root = "/Users/jeancharles/Desktop/Computer Vision-Qualification goals/Datasets/Loading Images with PyTorch Dataset", transform=transform)
```

```python
dataloader = torch.utils.data.DataLoader(dataset, batch_size=1, shuffle=True)
```

```python
images, labels = next(iter(dataloader))

# next(iter(dataloader)) returns a TUPLE made of a Tensor and a Single Number

# when we get images, we are getting the tensor of size torch.Size([3, 128, 128])

# by adding .permute(1,2,0), we are requesting a tensor of size torch.Size([128, 128, 3])

# We can now plot the image
plt.imshow(images[0].permute(1, 2, 0))
```
